```python
#importing libraries needed to access dataset file

import pandas as pd
import numpy as np
```

```python
#read the dataset

dt = pd.read_csv('fuel_ferc1.csv')
dt
```

```python
#finding the standard deviation and 75th precentile of fuel mmbtu per unit 
dt.fuel_mmbtu_per_unit.describe().round(2)
```

```python
#fuel type with the lowest average fuel cost per unit burned
dt.groupby('fuel_type_code_pudl').mean().round(2)
```

```python
#fuel type with the lowest average fuel cost per unit burned
dt.groupby('fuel_type_code_pudl').fuel_cost_per_unit_burned.mean()
```

```python
# year with the highest average fuel cost per unit delivered
highest_year = dt.groupby('report_year').fuel_cost_per_unit_delivered.mean()

highest_year.sort_values(ascending=False)
```

```python
#skewness and kurtosis of fuel qty burned
skewness = round(dt.fuel_qty_burned.skew(), 2)
print(skewness)
kurtosiss= dt.fuel_qty_burned.kurt().round(2)
kurtosiss
```

```python
#total number of missing values
print(dt.isnull().sum())

#percentage of missing values
dt.isnull().mean().round(4)*100
```

```python
#correlation
dt.corr().round(2)
```
